# NumPy ile Çok Boyutlu Dizi ve Matris İşlemleri

Bu çalışma, Python `numpy` kütüphanesi kullanılarak temel dizi işlemleri, boyut yönetimi, matris türleri ve istatistiksel hesaplamaların nasıl yapıldığını belgelemektedir.

```python
import numpy as np
```

## 1. Dizilere Giriş ve Veri Tipleri
Standart Python listelerinden NumPy dizilerine geçiş ve bellek yönetimi için `dtype` kullanımı.

```python
liste = [1, 2, 3, 4, 5]
dizi = np.array(liste)
print("Dizi:", dizi, "| Tipi:", type(dizi))

liste2 = [5, 2, 3, 4]
dizi2 = np.array(liste2)

# Bellek Optimizasyonu (Görüntü işleme için uint8 örneği)
resim = np.zeros((5, 5), dtype='uint8')
resim[2, 2] = 255
resim[2, 1] = 200
resim[1, 2] = 200
resim[2, 3] = 200
resim[3, 2] = 200
print("\nuint8 Resim Matrisi:\n", resim)
```

## 2. Otomatik Dizi Oluşturma Fonksiyonları
Belirli aralıklarda, adımlarla veya eşit parçalara bölünmüş diziler oluşturma (`arange`, `linspace`, `zeros`, `ones`, `full`).

```python
sifirmatris = np.zeros((2, 2))
birler = np.ones((3, 3))
full = np.full((3, 2), 12)

dizi_arange = np.arange(4, 10)
dizi_ondalik = np.arange(0, 1, 0.1)

dizi_linspace = np.linspace(0, 10, 5, endpoint=False)

print("Arange:", dizi_arange)
print("Ondalık Arange:", dizi_ondalik)
print("Linspace (Endpoint=False):", dizi_linspace)
```

## 3. Matris Türleri
Birim (`eye`), Köşegen (`diag`) ve Üçgen (`triu`, `tril`) matris yapıları.

```python
birim = np.eye(3)
print("Birim Matris:\n", birim)

matris = np.array([
    [2, 1, 3],
    [5, 4, 5],
    [1, 6, 4]
])
print("\nKöşegen (Diag):", np.diag(matris))

print("\nAlt Üçgen (Tril):\n", np.tril(matris))
print("\nÜst Üçgen K=1 (Triu):\n", np.triu(matris, k=1))
```

## 4. Şekil ve Boyut Manipülasyonu
Dizilerin boyutlarını değiştirme (`reshape`), tek boyuta indirme (`ravel`, `flatten`) ve yeni boyut ekleme (`newaxis`).

```python
dizi4 = np.array([5, 2, 3, 4]).reshape(2, 2)
print("Reshape (2x2):\n", dizi4)

dizi_tek_satir = np.ravel(matris)  # Orijinali etkiler
dizi_flatten = matris.flatten()    # Kopya oluşturur
print("\nRavel:", dizi_tek_satir)

dizi13 = np.array([1, 2, 3])
sutun_matrisi = dizi13[:, np.newaxis]
print("\nNewaxis (Sütun Matrisi):\n", sutun_matrisi)
```

## 5. Birleştirme ve Ayırma İşlemleri
Farklı dizileri dikey (`vstack`), yatay (`hstack`) veya eksen bazlı (`concatenate`) birleştirme ve bölme (`split`).

```python
ust = np.array([1, 2, 3])
alt = np.array([4, 5, 6])
print("Vstack:\n", np.vstack((ust, alt)))

sol = np.array([[10], [20]])
sag = np.array([[30], [40]])
print("\nHstack:\n", np.hstack((sol, sag)))

notlar = np.array([[50, 60], [70, 80], [90, 100], [40, 30]])
bolunmus = np.split(notlar, 2)
print("\nSplit:\n", bolunmus)
```

## 6. Matematiksel, İstatistiksel ve Koşullu İşlemler
Matematiksel fonksiyonlar, sınırlandırma (`clip`), istatistikler ve koşula bağlı değer atama (`where`).

```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])
print("Add:", np.add(a, b))
print("Power:", np.power(a, 2))

dizi = np.array([10, 50, 100, 200])
print("Clip (20-80 arası):", np.clip(dizi, 20, 80))

notlar = np.array([40, 70, 30, 90])
yeni_notlar = np.where(notlar < 50, 0, notlar)
print("Where (<50 ise 0 yap):", yeni_notlar)

veriler = np.array([10, np.nan, 20, np.inf, 30])
print("IsNaN kontrolü:", np.isnan(veriler))
```

## 7. Gelişmiş İndeksleme ve Filtreleme
Çok boyutlu dizilerde belirli elemanlara erişim ve boolean maskeleme ile veri filtreleme.

```python
np.random.seed(23)
numbers9 = np.random.randint(10, 50, 15).reshape(3, 5)
print("Ana Matris:\n", numbers9)

print("\nSütun/Satır Kesimi (Slicing):\n", numbers9[1:3, 1:4])

print("Maksimum değerin indeksi:", np.where(numbers9 == np.max(numbers9)))

filtrelenmis = numbers9[(numbers9 % 2 == 0) & (numbers9 > 20)]
print("20'den büyük çift sayılar:", filtrelenmis)
```
