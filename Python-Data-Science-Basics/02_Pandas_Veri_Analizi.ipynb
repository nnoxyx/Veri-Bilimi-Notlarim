# Pandas ile Veri Analizi ve Manipülasyonu

Bu çalışma, Python `pandas` kütüphanesi kullanılarak temel veri yapıları (Series ve DataFrame), veri filtreleme, gruplandırma (groupby), eksik veri yönetimi ve veri birleştirme (merge/concat) işlemlerinin nasıl yapıldığını belgelemektedir.

```python
import pandas as pd
import numpy as np
```

## 1. Pandas Serileri (Series)
Tek boyutlu veri yapısı olan Serilerin listelerden, sözlüklerden (dict) ve NumPy dizilerinden oluşturulması; indeks yapılandırmaları.

```python
random_numbers = np.random.randint(10, 100, 5)

# Özelleştirilmiş indekslerle Seri oluşturma
pandas_series = pd.Series([20, 30, 40, 50, 60], index=['a', 'b', 'c', 'd', 'e'])
print("Oluşturulan Seri:\n", pandas_series)

# İndeksleme ve Koşullu Seçim
print("\n'a' ve 'c' indeksli elemanlar:\n", pandas_series[['a', 'c']])
print("\n50'den büyük veya eşit olanlar:\n", pandas_series[pandas_series >= 50])
```

## 2. DataFrame Oluşturma ve Veri Seçimi (Indexing/Slicing)
İki boyutlu veri tablolarının oluşturulması, `loc` / `iloc` kullanımı ve sütun/satır düşürme (drop) işlemleri.

```python
df_rand = pd.DataFrame(np.random.randn(3, 3), index=['a', 'b', 'c'], columns=['Col1', 'Col2', 'Col3'])
print("Rastgele Verilerle DataFrame:\n", df_rand)

# Yeni Sütun Ekleme
df_rand['Col4'] = df_rand['Col1'] + df_rand['Col3']
print("\nYeni Sütun Eklendikten Sonra:\n", df_rand)

# Sütun Düşürme (Drop)
df_rand.drop('Col3', axis=1, inplace=True)
print("\nSütun Düşürüldükten Sonra (Col3 silindi):\n", df_rand)

# iloc ve loc ile Veri Çekme
print("\n'b' satırındaki Col1 ve Col2 verileri:\n", df_rand.loc['b', ['Col1', 'Col2']])
print("\nİndeks bazlı erişim (iloc[1, 1]):", df_rand.iloc[1, 1])
```

## 3. Veri Filtreleme ve Çoklu Koşullar
Mantıksal operatörler (`&`, `|`) ve `query()` metodu kullanılarak karmaşık filtreleme operasyonları.

```python
data_matrix = np.random.randint(10, 100, 75).reshape(15, 5)
df_filter = pd.DataFrame(data_matrix, columns=['C1', 'C2', 'C3', 'C4', 'C5'])

print("İlk 5 Satır:\n", df_filter.head())

# C1 sütununun 50'den büyük ve C2 sütununun çift sayı olduğu satırları seçme
kosul_sonucu = df_filter[(df_filter['C1'] >= 50) & (df_filter['C2'] % 2 == 0)]
print("\nKoşullu Filtreleme Sonucu:\n", kosul_sonucu[['C1', 'C2']])

# Query Kullanımı (Alternatif ve Okunabilir Yöntem)
query_sonucu = df_filter.query('C1 >= 50 & C2 % 2 == 0')
print("\nQuery Metodu ile Filtreleme Sonucu:\n", query_sonucu[['C1', 'C2']])
```

## 4. Gruplandırma (Groupby) İşlemleri
Kategorik verilere göre gruplandırma ve aggregasyon (toplam, ortalama, maksimum) hesaplamaları.

```python
personel_data = {
    'Çalışan': ['Ahmet Yılmaz', 'Can Ertürk', 'Hasan Korkmaz', 'Cenk Saymaz', 'Ali Turan', 'Rıza Ertürk', 'Mustafa Can'],
    'Departman': ['İK', 'Bilgi İşlem', 'Muhasebe', 'İK', 'Bilgi İşlem', 'Muhasebe', 'Bilgi İşlem'],
    'Yaş': [30, 25, 45, 50, 23, 34, 42],
    'Semt': ['Kadıköy', 'Tuzla', 'Maltepe', 'Tuzla', 'Kadıköy', 'Tuzla', 'Maltepe'],
    'Maaş': [5000, 3000, 4000, 3500, 2750, 6500, 4500]
}

df_personel = pd.DataFrame(personel_data)

print("Departmanlara Göre Toplam Maaş Gideri:\n", df_personel.groupby('Departman')['Maaş'].sum())
print("\nSemtlere Göre Ortalama Yaş:\n", df_personel.groupby('Semt')['Yaş'].mean())
print("\nDepartman Bazlı Çoklu Analiz:\n", df_personel.groupby('Departman')['Maaş'].agg(['mean', 'max', 'min']))
```

## 5. Eksik Veri (Missing Data) Analizi ve Temizleme
`NaN` (Not a Number) değerlerin tespiti, silinmesi (`dropna`) ve mantıksal değerlerle doldurulması (`fillna`).

```python
missing_data = np.random.randint(10, 100, 15).reshape(5, 3)
df_missing = pd.DataFrame(missing_data, index=['a', 'c', 'e', 'f', 'h'], columns=['Col1', 'Col2', 'Col3'])

# İndeksi genişleterek yapay eksik veriler (NaN) oluşturalım
df_missing = df_missing.reindex(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
print("Eksik Verili DataFrame:\n", df_missing)

print("\nHer sütundaki eksik veri sayısı:\n", df_missing.isnull().sum())

# Eksik verileri sütun ortalaması ile doldurma
df_filled = df_missing.fillna(df_missing.mean())
print("\nEksik Veriler Sütun Ortalamaları İle Doldurulduktan Sonra:\n", df_filled)
```

## 6. Veri Birleştirme (Merge ve Concat)
Farklı DataFrame'lerin ortak sütunlar (Join) veya eksenler (Concat) üzerinden birleştirilmesi.

```python
customers = {
    'CustomerId': [1, 2, 3, 4],
    'FirstName': ['Ahmet', 'Ali', 'Hasan', 'Canan']
}
orders = {
    'OrderId': [10, 11, 12, 13],
    'CustomerId': [1, 2, 5, 7],
    'OrderDate': ['2010-07-04', '2010-08-04', '2010-07-07', '2012-07-04']
}

df_cust = pd.DataFrame(customers)
df_ord = pd.DataFrame(orders)

print("İç Birleştirme (Inner Join - Sadece Kesişenler):\n", pd.merge(df_cust, df_ord, how='inner'))
print("\nSol Birleştirme (Left Join - Tüm Müşteriler):\n", pd.merge(df_cust, df_ord, how='left'))
print("\nDış Birleştirme (Outer Join - Her İki Taraftaki Tüm Kayıtlar):\n", pd.merge(df_cust, df_ord, how='outer'))
```

## 7. İleri Seviye Metotlar (Apply, String İşlemleri ve Pivot Tablolar)
Serilere veya DataFramelere fonksiyon uygulama (`apply`), metin manipülasyonları ve veri özetleme (`pivot_table`).

```python
data_adv = {
    'TextCol': ['Ahmet', 'Ali', 'Hasan', 'Ayşe', 'Burak'],
    'NumCol': [10, 20, 30, 40, 50]
}
df_adv = pd.DataFrame(data_adv)

# Lambda ile fonksiyon uygulama
df_adv['Kare'] = df_adv['NumCol'].apply(lambda x: x * x)
print("Apply Metodu Kullanımı:\n", df_adv)

# String metotları ile filtreleme ('A' ile başlayan veya içeren)
str_filter = df_adv[df_adv['TextCol'].str.lower().str.contains('a')]
print("\nİsminde 'a' harfi geçenler:\n", str_filter)

# Pivot Table Örneği
data_pivot = {
    'Ay': ['Mayıs', 'Haziran', 'Nisan', 'Mayıs', 'Haziran', 'Nisan', 'Mayıs'],
    'Kategori': ['Elektronik', 'Elektronik', 'Elektronik', 'Kitap', 'Kitap', 'Giyim', 'Giyim'],
    'Gelir': [20, 30, 15, 14, 32, 42, 12]
}
df_piv = pd.DataFrame(data_pivot)
print("\nPivot Tablo (Ay ve Kategoriye Göre Gelir Analizi):\n", df_piv.pivot_table(index='Ay', columns='Kategori', values='Gelir', aggfunc='sum'))
```
